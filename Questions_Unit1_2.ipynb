```python
from transformers import pipeline, set_seed, GPT2Tokenizer
```

```python
import os
import nltk
```

```python
set_seed(42)
```

```python
prompt = "The future of Artificial Intelligence is"
```

```python
# Initialize the pipeline with the specific model
bert_generator = pipeline('text-generation', model='bert-base-uncased')

# Generate text
output_bert = bert_generator(prompt, max_length=50, num_return_sequences=1)
print(output_bert[0]['generated_text'])
```

```python
roberta_generator = pipeline('text-generation', model='roberta-base')

output_roberta = roberta_generator(prompt, max_length=50, num_return_sequences=1)
print(output_roberta[0]['generated_text'])
```

```python
# Initialize the pipeline with the specific model
bart_generator = pipeline('text-generation', model='facebook/bart-base')

# Generate text
output_bart = bart_generator(prompt, max_length=50, num_return_sequences=1)
print(output_bart[0]['generated_text'])
```

```python
qa_bert = pipeline("question-answering", model="bert-base-uncased")
```

```python
questions = [
    "What are the risks?"
]

for q in questions:
    res = qa_bert(question=q, context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes.")
    print(f"\nQ: {q}")
    print(f"A: {res['answer']}")
```

```python
qa_roberta = pipeline("question-answering", model="roberta-base")
questions = [
    "What are the risks?"
]

for q in questions:
    res = qa_roberta(question=q, context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes.")
    print(f"\nQ: {q}")
    print(f"A: {res['answer']}")

```

```python
qa_bart = pipeline("question-answering", model="facebook/bart-base")
questions = [
    "What are the risks?"
]

for q in questions:
    res = qa_bart(question=q, context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes.")
    print(f"\nQ: {q}")
    print(f"A: {res['answer']}")

```

```python
masked_sentence = "The goal of Generative AI is to [MASK] new content."
mask_filler = pipeline("fill-mask", model="bert-base-uncased")
preds = mask_filler(masked_sentence)

for p in preds:
    print(f"{p['token_str']}: {p['score']:.2f}")
```

```python
mask_filler_roberta = pipeline("fill-mask", model="roberta-base")
masked_sentence_roberta = "The goal of Generative AI is to <mask> new content."
#[MASK] changed to <mask> as roberta needs in "<mask>" format, else throws error
preds = mask_filler_roberta(masked_sentence_roberta)

for p in preds:
    print(f"{p['token_str']}: {p['score']:.2f}")
```

```python
mask_filler_bart = pipeline("fill-mask", model="facebook/bart-base")
masked_sentence = "The goal of Generative AI is to <mask> new content."
#[MASK] changed to <mask> as roberta needs in "<mask>" format, else throws error
preds = mask_filler_bart(masked_sentence)

for p in preds:
    print(f"{p['token_str']}: {p['score']:.2f}")
```
